#### %>% 연산자를 이용한 연속 처리
 - %>% 연산자를 사용해 일련의 가공 작업을 연결

<mark>chain(pipe) operator (%>%)</mark> : 여러 단계 절차가 필요해서 중간결과 저장 후 그 객체를 후속 절차에서 받아 사용해야 하는 경우에 유용함.

```r
# 데이터 셋 읽어들이기 : avocado.csv
# Date, AveragePrice : 평균 단가 , Total Volume : 판매량 etc...

# 데이터 셋을 읽어들여 str 함수를 이용해 프레임의 구성 살펴보기
avocado <- read.csv("avocado.csv", header = TRUE ,sep = ",")
str(avocado)
```

#### 그룹 단위 통계(1)
 - 경출 도출을 위해 총 판매량과 판매 가격 속성을 지역에 따라 구분하여 각각 요약
 - dplyr 라이브러리의 group_by 와 summarize 함수 사용.

```r
((x_avg = avocado %>% group_by(region) %>% 
 summarize(V_avg = mean(Total.Volume), P_avg = mean(AveragePrice))))
```

#### 그룹 단위 통계(2)
 - 지역별 특징을 연도별로 다시 세분화

```r
(x_avg = avocado %>% group_by(.,region, year) %>% 
 summarize(V_avg = mean(Total.Volume), P_avg = mean(AveragePrice)))
```

#### 그룹 단위 통계(3)
 - 유기농 여부(type) 를 기준으로 한 번 더 세분화

```r
(x_avg = avocado %>% group_by(region, year, type) %>% 
summarize(V_avg = mean(Total.Volume), P_avg = mean(AveragePrice)))
```

#### 그룹 단위 통계(4)
 - 방대한 샘플 데이터로부터 지역, 연도, 유기농 재배 여부를 기준으로
 총 판매량과 평균 가격의 요약된 통계를 얻음
 - 6장에서 시각화를 배움

### Date 형 데이터의 활용(1)
 - Date형 속성은 1개월은 31일 , 1년은 12개월로 구성되어 일반 숫자형처럼 처리할 경우 데이터 간의 간격이 일정하지 않아 시각화나 모델링 단계에서 잘못 적욛될 수 있으므로 특별 처리 필요.
 - <mark>속성 하나에 세 가지 속성(연-월-인) 정보를 포함하는 것</mark>이므로 적절히 가공하여 활용하면 데이터를 좀 더 세밀하게 분석할 수 있음.

 - avocdo 판매 정보를 이번에는 연도별 평균 대신 월별 평균으로 요약
 - Date 형 속성인 Date에서 month를 추출할려면 lubridate 라이브러리에서 제공하는 month 함수 사용. month() <mark>year나 day 함수도</mark> 함수도 활용 가능

```r
library(lubridate)
```

```r
(x_avg = avocado %>% group_by(region, year, month(Date), type) %>% 
 summarize(V_avg = mean(Total.Volume), P_avg = mean(AveragePrice)))
```

#### 두 개의 데이터 프레임을 병함(1)
 - 이제까지 봐왔던 일반 데이터 프레임의 구성
   - 속성 하나를 열 하나에 배치하고
   - 행 하나가 샘플 한 개의 기록을 구성과 다름
 - <mark> 각 측정 연도 값이 열의 이름이 되어 한 행에 여러 데이터가 기록됨.</mark>
 - <mark> 국가 명, 연도, 전기 생산량과 소비량을 가각 하나의 열에 대응시키는 변형이 필요.
 - gather 함수

```r
library(tidyr)
```

```r
elec_gen_df = gather(elec_gen, -country, key= "year", value="ElectricityGeneration")
elec_use_df = gather(elec_use, -country, key="year", value = "ElectricityUse")
```

### 두 개의 데이터 프레임을 병합

 - <mark>merge()</mark>는 두 데이터 프레임을 공통된 값을 기준으로 묶는 함수이다.
 - 데이터 베이스에서 join 과 같은 역할을 한다.
